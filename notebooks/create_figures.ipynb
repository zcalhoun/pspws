# Create figures
This notebook contains the code to generate the figures used in the paper.

```python
import numpy as np
import geopandas as gpd
from matplotlib import pyplot as plt
```

```python
# Load any of the files that you want to look at.
data = gpd.read_file('../preprocessed_data/charlotte_july_2023/census_tract.shp')

# Remove any rows with population less than 0
data = data[data['population'] > 0]

# set the crs
data.crs = "EPSG:4326"

# project to a local crs
data = data.to_crs("EPSG:32617")
```

```python
# We need these to set the vmax and vmin parameters for the plot
mu = data['m1_temp'].mean()
std = data['m2_temp'].std()
```

```python
print("Model 1 results")

data.plot(column='m1_temp', edgecolor='black', legend=True, figsize=(8,5), cmap='coolwarm', legend_kwds={'shrink':0.8}, vmax=mu+2*std, vmin=mu-2*std)
plt.axis('off')
plt.show()
```

```python
print("Model 2 results")
data.plot(column='m2_temp', edgecolor='black', legend=True, figsize=(8,5), cmap='coolwarm', legend_kwds={'shrink':0.8}, vmax=mu+2*std, vmin=mu-2*std)
plt.axis('off')
plt.show()
```

```python
data['diff'] = data['m2_temp'] - data['m1_temp']

print("Model 2 and model 1 difference")
data.plot(column='diff', edgecolor='black', legend=True, figsize=(8,5), cmap='coolwarm', legend_kwds={'shrink':0.8}, vmax=1, vmin=-1)
plt.axis('off')
plt.show()
```

```python
# Double check the column names -- eve_avg, tempAvg, and temp are all
# column names in the data.
print("Observed areal groups")
data.plot(column='temp', edgecolor='black', legend=True, figsize=(8,5), cmap='coolwarm', legend_kwds={'shrink':0.8}, missing_kwds={'color':'lightyellow', 'hatch':'//////', 'edgecolor':'lightgrey'},vmax=mu+2*std, vmin=mu-2*std)
plt.axis('off')
plt.show()
```
