# Graphical Table of Contents
This notebook contains the code used to construct the graphical table of contents.

```python
import numpy as np
from matplotlib import pyplot as plt
import geopandas as gpd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = gpd.read_file('../preprocessed_data/durham_july_2021/durham_july_2021_block_group.shp')

# Need to create an intensity column for the graphs we want to make
df['intensity'] = df['num_statio'] / df['population'] * 1000
```

```python
# Create table of contents graphic
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(3.15, 1.57), width_ratios=[1,1,1], height_ratios=[29, 1], layout='tight')

df.plot('intensity', ax=ax1, legend=True, cmap='bone_r', legend_kwds={"cax": ax4, 'orientation':'horizontal', 'extend':'max', 'ticks': [0, 1, 2, 3]}, edgecolor='lightgray', linewidth=0.1, vmax=3)
df.plot('m1_temp', ax=ax2, legend=True, cmap='coolwarm', legend_kwds={ "cax": ax5, 'orientation':'horizontal', 'extend':'both', 'drawedges':False}, edgecolor='lightgray', linewidth=0.1, vmin=27, vmax=29)
df.plot('m2_temp', ax=ax3, legend=True, cmap='coolwarm', legend_kwds={"cax": ax6 , 'orientation':'horizontal', 'extend':'both'}, edgecolor='lightgray', linewidth=0.1, vmin=27, vmax=29)
ax1.axis('off')
ax2.axis('off')
ax3.axis('off')
ax1.set_title("Stations per\n1000 people ", fontsize='xx-small', pad=0)
ax2.set_title("Uncorrected\nestimate ($^{\circ}C$)", fontsize='xx-small',pad=0)
ax3.set_title("Corrected\nestimate($^{\circ}C$)", fontsize='xx-small',pad=0)

# Make font on colorbars smaller
ax4.set_xticklabels(['0', '1', '2', '3'], fontsize='xx-small')
ax5.set_xticklabels(['27', '28', '29'], fontsize='xx-small')
ax6.set_xticklabels(['27', '28', '29'], fontsize='xx-small')

# Save as pdf
# plt.savefig('../results/figures/table_of_contents.pdf', bbox_inches='tight', dpi=300)
plt.show()
```
